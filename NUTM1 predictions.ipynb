```python
transcript_ids_to_protein_sequences = {}
transcript_id_to_gene_name = {}
for gene in genome.genes():
    for t in gene.transcripts:
        if t.complete:
            p = t.protein_sequence
            if p:
                transcript_id_to_gene_name[t.id] = t.gene.name
                transcript_ids_to_protein_sequences[t.id] = p
```

```python
import progressbar
pbar = progressbar.ProgressBar()
kmers_to_gene_names = defaultdict(list)
for tid, p in pbar(transcript_ids_to_protein_sequences.items()):
    gene_name = transcript_id_to_gene_name[tid]
    n = len(p)
    for k in range(8, 12):
        for i in range(n - k + 1):
            kmer = p[i:i+k]
            kmers_to_gene_names[kmer].append(gene_name)
kmers_to_unique_gene_names = {kmer: set(gene_names) for (kmer, gene_names) in kmers_to_gene_names.items()}
```

```python
from collections import defaultdict
import pandas as pd
netmhcpan_data = defaultdict(list)
cols = None
float_cols = {"Aff(nM)", "%Rank_BA", "%Rank_EL", "Score_EL"}
int_cols = {"Pos", "Of", "Gp", "Gl", "Ip", "Il"}
with open("netmhcpan4.1-nutm1.txt") as f:
    for l in f:
        l = l.strip()
        if l and not l.startswith("#"):
            parts = l.split()

            if cols is None:
                cols = parts
            else:
                if len(parts) == len(cols) - 1:
                    parts.append("")
                elif len(parts) == len(cols) + 1:
                    parts = parts[:-2] + parts[-1:]
                assert len(parts) >= len(cols), (parts, cols)
                for (k,v) in zip(cols, parts):
                    if k in float_cols: 
                        v = float(v)
                    elif k in int_cols:
                        v = int(v)
                    netmhcpan_data[k].append(v)
df_netmhcpan = pd.DataFrame(netmhcpan_data)
```

```python
from collections import Counter
allele_counts = Counter()
chosen_peptides = []
desired_allele_coverage = {
    a: 2 for a in set(df_netmhcpan.MHC)
}
desired_allele_coverage["HLA-A*02:01"] += 1

peptide_to_best_netmhcpan_percentile_rank = {}
for p, subdf in df_netmhcpan.groupby("Peptide"):
    peptide_to_best_netmhcpan_percentile_rank[p] = subdf["%Rank_EL"].min()
    

for p, subdf in sorted(df_netmhcpan_subset.groupby("Peptide"), key=lambda x: peptide_to_best_netmhcpan_percentile_rank[x[0]]):

    # always keep mass spec validated peptides
    keep = p in mass_spec_confirmed_not_in_paralogs
    for a in subdf.MHC:
        allele_counts[a] += 1
        if allele_counts[a] <= desired_allele_coverage[a]:
            keep = True
    if keep:
        print("Keeping %s (%d alleles: %s)" % (p, len(subdf), set(subdf.MHC)))
        print(subdf[["MHC", "Score_EL",  "%Rank_EL",  "Score_BA",  "%Rank_BA",  "Aff(nM)"]])
        chosen_peptides.append(p)
    else:
        print(" -- Skipping %s (%d redundant alleles)" % (p, len(subdf)))
```
